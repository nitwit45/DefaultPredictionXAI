```python
features = df.columns.to_list()
X_features = features[:-1]
```

```python
X = df.drop(["default.payment.next.month"], axis = 1)
y = df["default.payment.next.month"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
df_X_train = pd.DataFrame(data=X_train, columns=X_features)
df_X_test = pd.DataFrame(data=X_test, columns=X_features)
df_y_train = pd.DataFrame(data=y_train, columns=['default.payment.next.month'])
df_y_test = pd.DataFrame(data=y_test, columns=['default.payment.next.month'])
```

```python
df_X_train.to_csv('df_X_train.csv', index=False)
df_X_test.to_csv('df_X_test.csv', index=False)
df_y_train.to_csv('df_y_train.csv', index=False)
df_y_test.to_csv('df_y_test.csv', index=False)
```

```python
import xgboost as xgb

model = xgb.XGBClassifier(n_estimators = 400, max_depth = 4, scale_pos_weight = 3, min_child_weight = 4, reg_alpha = 100, subsample = 0.64, colsample_bytree = 0.55, random_state = 42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn.metrics import classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
# Plot the confusion matrix using seaborn and matplotlib
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Print the classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
import lime
import lime.lime_tabular
```

```python
explainer = lime.lime_tabular.LimeTabularExplainer(X_train.values, mode='classification', feature_names=X_train.columns)
```

```python
#Choose a specific instance for explanation (e.g., the first instance in the test set)
instance = X_test.values[0]
```

```python
# Generate explanations using LIME
num_features = X_test.shape[1]
lime_explanation = explainer.explain_instance(instance, model.predict_proba, num_features=num_features)
```

```python
# Display the LIME explanation
lime_explanation.show_in_notebook()
```
